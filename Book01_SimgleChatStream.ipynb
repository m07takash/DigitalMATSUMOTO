```python
import DigiM_Agent as dma
```

```python
agent_file = "agent_01DigitalMATSUMOTO_GPT.json"
agent = dma.DigiM_Agent(agent_file)
```

```python
query = "最近の日本企業ってどう思いますか？"
```

```python
def test_message(agent, query, stream_mode):
    response = ""
    for prompt, response_chunk, completion in agent.generate_response("LLM", query, stream_mode=stream_mode):
        if response_chunk:
            response += response_chunk
            yield response_chunk

    print("")
    print("[最終的な出力]")    
    print(response)
```

```python
stream_mode=True
for response in test_message(agent, query, stream_mode):
    print(response)
```
