```python
import os
from dotenv import load_dotenv
import chromadb

# system.envファイルをロードして環境変数を設定
load_dotenv("system.env")
rag_mst_file = os.getenv("RAG_MST_FILE")
rag_folder_db_path = os.getenv("RAG_FOLDER_DB")

# ChromaDBクライアント
db_client = chromadb.PersistentClient(path=rag_folder_db_path)
```

```python
# すべてのコレクションを取得
collections = db_client.list_collections()

# コレクション一覧を表示
print("=== コレクション一覧 ===")
for col in collections:
    print(f"- {str(col)}")
```

```python
import pandas as pd

# すべてのコレクションをループ
for col_name in ["DigiMATSU_Opinion_Insight"]: #collections:
    col_name_str = str(col_name)  # コレクション名を取得
    collection = db_client.get_collection(name=col_name_str)  # コレクションを取得
    records = collection.get()  # すべてのデータを取得

    # データを DataFrame に変換
    df = pd.DataFrame({
        "id": records.get("ids", []),
        "document": records.get("documents", []),
        "metadata": records.get("metadatas", []),
        "embedding": records.get("embeddings", [])
    })

    # 出力
    print(f"\n=== コレクション: {col_name_str} ===")
    print(df)

    # CSV に保存（オプション）
#    df.to_csv(f"{col_name_str}_records.csv", index=False)
```
